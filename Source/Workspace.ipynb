```python
import numpy as np
```

```python
import cv2 as cv
```

```python
import glob
```

```python
__proj_dir__ = "E:\Projects\dev\LiRead"
```

```python
img8bits = []
```

```python
for filename in glob.glob(__proj_dir__ + "/Resources/*.PNG"): img8bits.append(cv.imread(filename))
```

```python
img8bits = np.array(img8bits, dtype = tuple)
```

```python
X = []
for img8bit in img8bits:
    coloravg = [ 0, 0, 0 ]
    for row in img8bit:
        for pixel in row:
            coloravg[0] += pixel[0] 
            coloravg[1] += pixel[1] 
            coloravg[2] += pixel[2]
    X.append([ coloravg[0] * 3 / img8bit.size, coloravg[1] * 3 / img8bit.size, coloravg[2] * 3 / img8bit.size ])
```

```python
X = [ [ x[0] / 255, x[1] / 255, x[2] / 255 ] for x in X ]
```

```python
X
```

```python
brightness = [ np.clip(np.linalg.norm(x), 0, 1) for x in X ]
plt.plot(brightness)
plt.show()
```
